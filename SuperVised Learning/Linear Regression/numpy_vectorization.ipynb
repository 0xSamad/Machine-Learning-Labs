# Python , Numpy and Vectorization

```python
import numpy as np
import time
```

## Vector Creation

```python
a = np.zeros(4)
print(f"np.zeros(4) :   a = {a}, a shape = {a.shape}, a data type = {a.dtype}")
a = np.zeros((4,))
print(f"np.zeros(4,) :  a = {a}, a shape = {a.shape}, a data type = {a.dtype}")
a = np.random.random_sample(4)
print(f"np.random.random_sample(4): a = {a}, a shape = {a.shape}, a data type = {a.dtype}")
```

```python
a = np.arange(4.)           
print(f"np.arange(4.): a = {a}, a shape = {a.shape}, a data type = {a.dtype}")
a = np.random.rand(4)         
print(f"np.random.rand(4): a = {a}, a shape = {a.shape}, a data type = {a.dtype}")
```

# Indexing
Indexing means referring to an element of an array by its position within the array.

NumPy starts indexing at zero so the 3rd element of an vector  𝐚
  is a[2].

```python
a = np.arange(10)
print(a)

# access an element
print(f"a[2].shape : {a[2].shape} a[2]= {a[2]}")

# access an element with a negative index

print(f"a[-1]= {a[-1]}")

# indexes must be within the range of the vector or they will produce an error

try:
    print(a[10])
except Exception as e:
    print("The error message you will see is: ")
    print(e)
```

### 3.4.2 Slicing
Slicing creates an array of indices using a set of three values (`start:stop:step`). A subset of values is also valid. Its use is best explained by example:

```python
a = np.arange(10)
print(f" a = {a}")

# access 5 consecutive elements
print(f"a[1,6,1] = {a[1:6:1]}")

# access 3 elements separted by 2
c = a[2:7:2]
print("a[2:7:2] = ", c)

# access all elements index 3 and above
c = a[3:]
print("a[3:]    = ", c)

# access all elements below index 3
c = a[:3]
print("a[:3]    = ", c)

# access all elements
c = a[:]
print("a[:]     = ", c)
```

## Single Vector Operations

```python
a = np.array([1,2,3,4])

b = -a
print(b)

# sum
# sum all the elements in the array and returns a scalar
c = np.sum(a)
print(c)

# mean

d = np.mean(a)
print(d)

# power

e = a **2
print(e)
```

## Vector dot Product

```python
a = np.array([1, 2, 3, 4])
b = np.array([-1, 4, 3, 2])
c = np.dot(a, b)
print(f"NumPy 1-D np.dot(a, b) = {c}, np.dot(a, b).shape = {c.shape} ") 
c = np.dot(b, a)
print(f"NumPy 1-D np.dot(b, a) = {c}, np.dot(a, b).shape = {c.shape} ")
```

## The need for Speed : Vector Vs loop

```python
np.random.seed(1)
import time
a = np.random.rand(1000000)
b = np.random.rand(1000000)
start = time.time()
c = np.dot(a,b)
end = time.time()
print(f"Total time taken by vectorization: {end-start:.4f}")

start = time.time()
total = 0
for i in range(1000000):
    c = a[i]*b[i]
    total+=c
end = time.time()
print(f"Total time taken by loop = {end - start:.4f}")
```

## Matrix Creation

```python
a = np.zeros((1, 5))                                       
print(f"a shape = {a.shape}, a = {a}")                     

a = np.zeros((2, 1))                                                                   
print(f"a shape = {a.shape}, a = {a}") 

a = np.random.random_sample((1, 1))  
print(f"a shape = {a.shape}, a = {a}") 
```

## Matrix Indexing

```python
#vector indexing operations on matrices
a = np.arange(6).reshape(-1, 2)   #reshape is a convenient way to create matrices
print(f"a.shape: {a.shape}, \na= {a}")

#access an element
print(f"\na[2,0].shape:   {a[2, 0].shape}, a[2,0] = {a[2, 0]},     type(a[2,0]) = {type(a[2, 0])} Accessing an element returns a scalar\n")

#access a row
print(f"a[2].shape:   {a[2].shape}, a[2]   = {a[2]}, type(a[2])   = {type(a[2])}")
```

## Slicing


```python
#vector 2-D slicing operations
a = np.arange(20).reshape(-1, 10)
print(f"a = \n{a}")

#access 5 consecutive elements (start:stop:step)
print("a[0, 2:7:1] = ", a[0, 2:7:1], ",  a[0, 2:7:1].shape =", a[0, 2:7:1].shape, "a 1-D array")

#access 5 consecutive elements (start:stop:step) in two rows
print("a[:, 2:7:1] = \n", a[:, 2:7:1], ",  a[:, 2:7:1].shape =", a[:, 2:7:1].shape, "a 2-D array")

# access all elements
print("a[:,:] = \n", a[:,:], ",  a[:,:].shape =", a[:,:].shape)

# access all elements in one row (very common usage)
print("a[1,:] = ", a[1,:], ",  a[1,:].shape =", a[1,:].shape, "a 1-D array")
# same as
print("a[1]   = ", a[1],   ",  a[1].shape   =", a[1].shape, "a 1-D array")
```
