# Predict Car prices based on age in years

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
age_train = np.array([1.0,2,3,4])
price_train = np.array([45.0,40,36,31])

print(f"Shape of age_train {age_train.shape}")
print(f"Shape of age_train {price_train.shape}")

m = age_train.shape[0]
```

```python
i =3
x_i = age_train[i]
y_i = price_train[i]
print(np.c_[x_i,y_i])
```

## plot the raw data

```python
plt.scatter(age_train,price_train,marker='x',c='r')
plt.title("Car price prediction")
plt.xlabel("Age of car in years")
plt.ylabel("price of car in 1000$")
plt.show()
```

```python
def compute_model(x,w,b):
    n = x.shape[0]
    f_wb = np.zeros(n)
    for i in range(n):
        f_wb[i] = w * x[i] + b
    return f_wb
```

```python
w = -4.77
b = 50

y_hat = compute_model(age_train,w,b)
plt.scatter(age_train,price_train,marker='x',c='b',label='real values')
plt.plot(age_train,y_hat,c='r',label='our prediction')
plt.xlabel("Age of car in years")
plt.ylabel("price of car in 1000$")
plt.legend()
plt.show()
```
